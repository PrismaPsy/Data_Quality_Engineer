```python
# Importar las librerías correspondientes.
import pandas as pd
```

```python
dataset_to_transform = pd.read_json('../Datasets/taylor_swift_spotify.json')
```

```python
# Abrir el nuevo arhivo de extensión CSV en un nuevo DataFrame.
dataset = pd.read_csv('../Datasets/dataset.csv')

# Visualizar el DataFrame nuevamente para asegurarnos que la información se haya guardado correctamente.
dataset.head()
```

```python
# Visualizar el tipo de dato que contiene cada columna en el DataFrame.
dataset.dtypes
```

```python
# Revisar datos nulos en el DataFrame.
dataset.isnull()
```

```python
# Visualizar la columna "Albums" para ver su contenido.
dataset['albums'].head()
```

```python
# Visualizar con más detalle la columna "Albums".

# Mostrar los datos anidados de la columna sin desanidar completamente.
def print_nested_data(row):
    print(row['albums'])

# Aplicar la función a cada fila de la columna.
dataset.apply(print_nested_data, axis = 1)
```

```python
# Desanidar los datos de la columna "Albums" a partir del archivo con la extensión JSON, 
# usando la función 'json_normalize' y guardando su contenido en un nuevo DataFrame.
data_albums = pd.json_normalize(dataset_to_transform['albums'])
```

```python
# Visualizar el nuevo DataFrame para entender su contenido.
data_albums.head()
```

```python
# Mostrar los últimos registros del DataFrame.
data_albums.tail(5)
```

```python
# Visualizar el tipo de datos de cada columna,
data_albums.dtypes
```

```python
# Revisar datos nulos en el nuevo DataFrame.
data_albums.isnull()
```

```python
# Sumar los nulos para verificar.
data_albums.isnull().sum()
```

```python
# Revisar datos duplicados.

# Buscar duplicados en todas las columnas, excepto en 'Tracks', ya que contiene datos anidados.
column_ignore = 'tracks'
column_considere = data_albums.columns[data_albums.columns != column_ignore]

# Eliminar temporalmente la columna 'Tracks' para identificar duplicados.
duplicate_registres = data_albums[data_albums.drop(column_ignore, axis = 1).duplicated(subset = column_considere, keep = False)]
duplicate_registres
```

```python
# Visualizar la columna "Tracks" para ver su contenido.
data_albums['tracks'].head()
```

```python
# Visualizar con más detalle la columna "Tracks".

# Mostrar los datos anidados de la columna sin desanidar completamente.
def print_nested_data(row):
    print(row['tracks'])

# Aplicar la función a cada fila de la columna.
data_albums.apply(print_nested_data, axis = 1)
```

```python
# Explotar la columna "Tracks" para expandir sus elementos en filas separadas
data_expanded = data_albums.explode('tracks')
```

```python
# Crear un nuevo DataFrame con los datos expandidos.
data_tracks = pd.json_normalize(data_expanded['tracks'])
```

```python
# Visualizar el nuevo DataFrame para analizar su contenido.
data_tracks.head()
```

```python
# Copiar la información del DataFrame de 'Tracks' en un DataFrame nuevo,
# para reducirlo a las columnas necesarias.
data_tracks_only = data_tracks
delete_t = ['audio_features.danceability', 'audio_features.energy', 'audio_features.key', 'audio_features.loudness',
          'audio_features.mode', 'audio_features.speechiness', 'audio_features.acousticness', 
          'audio_features.instrumentalness', 'audio_features.liveness', 'audio_features.valence',
          'audio_features.tempo', 'audio_features.id', 'audio_features.time_signature']
data_tracks_only = data_tracks_only.drop(delete_t, axis = 1)
```

```python
# Visualizar el nuevo DataFrame.
data_tracks_only.head()
```

```python
# Analizar tipo de datos en las columnas del nuevo DataFrame.
data_tracks_only.dtypes
```

```python
# Revisar los datos nulos.
data_tracks_only.isnull().sum()
```

```python
# Revisar los duplicados por columna. 
duplicates_columns_t = data_tracks_only.apply(lambda x: x.duplicated().sum())
duplicates_columns_t
```

```python
# Copiar la información del DataFrame de 'Tracks' en un DataFrame nuevo,
# para reducirlo a las columnas necesarias.
data_audio_features = data_tracks
delete_f = ['disc_number', 'duration_ms', 'explicit', 'track_number', 'track_popularity', 'track_name']
data_audio_features = data_audio_features.drop(delete_f, axis = 1)
```
